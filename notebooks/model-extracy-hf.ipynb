```python
import requests
import json

# Define the API endpoint
api_url = "https://huggingface.co/models-json?pipeline_tag=text-classification&p={page}&sort=likes&withCount=false"

for p in range(3):
    response = requests.get(api_url.format(page=p))
    if response.status_code == 200:
        data = response.json()['models']
        markdown_list = "\n".join([f"- {model['id']}" for model in data])
        print(markdown_list)        
    else:
        print(f"Failed to retrieve data: {response.status_code}")
```
