```python
from datasets import load_dataset

ds = load_dataset("SetFit/20_newsgroups")
```

```python
ds
```

```python
train = ds["train"].to_pandas()
test = ds["train"].to_pandas()
dfx = train.merge(test)
dfx['label_encoded'] = dfx['label']
```

```python
dfx.loc[dfx[dfx['text'] == ''].index,"text"] = None
```

```python
dfx.isnull().sum() 
```

```python
dfx.dropna(subset=['text'], inplace=True)
```

```python
dfx.to_csv('../data/processed/20_newsgroups.csv',index=False)
```
